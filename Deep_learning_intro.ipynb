```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_regression

data=make_regression(n_samples=70,n_features=1,noise=10, random_state=42)
```

```python
data[0][:5]
```

```python
data[1][:5]
```

```python
X=data[0]
y=data[1]
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X,y)
plt.show()
```

```python
# Simple way (classic way) of resolving this issue

def classic_way(d): return d*60

y_pred=[classic_way(d) for d in X]

plt.figure(figsize=(8,5))
plt.scatter(X,y)
plt.plot(X,y_pred, c='red')
plt.show()
```

**ML** way

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(X,y)

y_pred=lr_model.predict(X)

plt.figure(figsize=(8,5))
plt.scatter(X,y)
plt.plot(X,y_pred, c='red')
plt.xlabel('X')
plt.ylabel('y')
equ=f"y={lr_model.coef_[0]:.2f}x + {lr_model.intercept_:.2f}"
plt.title(equ)
plt.show()
```

**Deep Learning Way**

```python
from sklearn.datasets import make_regression

data=make_regression(n_samples=70,n_features=1,noise=10, random_state=42)
X=data[0]
y=data[1]

from sklearn.neural_network import MLPRegressor
dl_model=MLPRegressor(max_iter=2000)
dl_model.fit(X,y)
y_pred=dl_model.predict(X)

plt.figure(figsize=(8,5))
plt.scatter(X,y)
plt.plot(X,y_pred, c='red')
plt.xlabel('X')
plt.ylabel('y')

plt.show()
```
